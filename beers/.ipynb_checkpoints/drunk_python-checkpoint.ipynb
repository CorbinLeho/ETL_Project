```python
#Dependencies
import pandas as pd
```

```python
path = "open-beer-database-breweries.csv"
csv_breweries = pd.read_csv(path, sep=";")
print(csv_breweries)
```

```python
breweries_df = pd.DataFrame(csv_breweries)
breweries_df.head()
```

```python
breweries_df = breweries_df.rename(columns={"id": "brewery_id","state":"STATE"})
breweries_df.head()
```

```python
#Google search says id 812 should be "magic hat brewery" for beer 'hocus pocus'
check = breweries_df[breweries_df["brewery_id"]==812]['breweries']
print(check)
```

```python
path = "beers-fork.csv"
csv_beers = pd.read_csv(path, sep=",")
print(csv_beers)
```

```python
beers_df = pd.DataFrame(csv_beers)
beers_df.head()
```

```python
booze_df = pd.merge(breweries_df,beers_df,on="brewery_id")
booze_df.head()
```

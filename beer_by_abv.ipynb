```python
import pandas as pd
import numpy as np
```

```python
file = "beer_reviews.csv"
file2 = "breweries.csv"
```

```python
reviews_df = pd.read_csv(file)
breweries_df = pd.read_csv(file2)
```

```python
breweries_df = breweries_df.rename(columns = {"name":"brewery_name"})
breweries_df.head()
```

```python
merge_brewery_df = pd.merge(breweries_df, reviews_df, on = "brewery_name")
merge_brewery_df.head()
```

```python
merge_brewery_df_2 = merge_brewery_df.drop_duplicates("beer_name")
merge_brewery_df_2.head()
```

```python
grouped_brewery_df = merge_brewery_df_2.groupby(["state"])
grouped_brewery_df.head()
```

```python
average_abv = grouped_brewery_df["beer_abv"].mean().round(1)
average_abv
```

```python
abv_df = pd.DataFrame(average_abv)
abv_df.sort_values(by = "beer_abv", ascending = False, inplace = True)
abv_df
```

```python
abv_df.to_csv("ETLOutput/abv_by_state.csv", header = True)
```

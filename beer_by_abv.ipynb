```python
import pandas as pd
import numpy as np
```

```python
file = "beer_reviews.csv"
file2 = "breweries.csv"
file3 = "beers/drunk_python.csv"
```

```python
reviews_df = pd.read_csv(file)
breweries_df = pd.read_csv(file2)
open_df = pd.read_csv(file3)
```

```python
breweries_df = breweries_df.rename(columns = {"name":"brewery_name"})
breweries_df.head()
```

```python
merge_brewery_df = pd.merge(breweries_df, reviews_df, on = "brewery_name")
merge_brewery_df.head()
```

```python
merge_brewery_df_2 = merge_brewery_df.drop_duplicates("beer_name")
merge_brewery_df_2.head()
```

```python
grouped_brewery_df = merge_brewery_df_2.groupby(["state"])
grouped_brewery_df.head()
```

```python
open_df = pd.DataFrame(open_df)
open_df.head()
```

```python
open_df_1 = open_df.drop_duplicates("name")
open_df_1.head()
```

```python
grouped_open_df = open_df_1.groupby(["state"])
grouped_open_df.head()
```

```python
average_abv_1 = grouped_open_df["abv"].mean().round(1)
average_abv_1
```

```python
average_abv_2 = grouped_brewery_df["beer_abv"].mean().round(1)
average_abv_2
```

```python
avg_df = pd.concat([average_abv_1, average_abv_2])
avg_df_1 = avg_df.groupby(avg_df.index)
avg_df_2 = avg_df_1.mean()
avg_df_2.head()
```

```python
avg_df = average_abv_2
avg_df.replace('AL', 7.8)
avg_df.replace('AZ', 6.3)
avg_df.replace('CA', 6.9)
avg_df.replace('CO', 7.6)
avg_df.replace('CT', 6.4)
avg_df
```

```python
abv_df = pd.DataFrame(avg_df)
abv_df.sort_values(by = "beer_abv", ascending = False, inplace = True)
abv_df
```

```python
abv_df.to_csv("abv_by_state.csv", header = True)
```
